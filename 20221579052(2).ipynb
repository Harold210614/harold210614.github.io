```python
import dash
from dash import html, dcc,dash_table
import dash_html_components as html
import dash_core_components as dcc
import dash_table
import dash_bootstrap_components as dbc
import folium
import geopandas as gpd
import pandas as pd
```

```python
app = dash.Dash(__name__)
map = folium.Map(location=[3.74590, -71.78598], zoom_start=11)
tile = folium.TileLayer(
    tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}', attr = 'Esri', name = 'Esri Satellite', overlay = False, control = True).add_to(map)

folium.Marker(
    location=[3.747036, -71.866334],
    tooltip="PENDARE",
    popup="Timberline Lodge",
    icon=folium.Icon(color="black"),
).add_to(map)

def style_ruta_1(feature):
  return{
      'color': 'RGB(000,255,255)',
      'weight': 4,
      'dashArray': '8, 8',
  }

ruta_1 = gpd.read_file('/content/drive/MyDrive/ruta_1.zip')
folium.GeoJson(ruta_1,style_function=style_ruta_1).add_to(map)


def style_fuction_2(feature):
  return{
      'color': 'RGB(255,0,255)',
      'weight': 4,
      'dashArray': '8, 8',
  }

ruta2 = gpd.read_file('/content/drive/MyDrive/Programación/ruta2.zip')
folium.GeoJson(ruta2,style_function=style_fuction_2).add_to(map)

data = {
    'Ruta':[1,2],
    'Longitud Sabana (Km)':[4.2,3.1],
    'Longitud Vía (Km)':[11.9,14.7],
    'PHD (UND)':[2,1],
    'Longitud Total (Km)':[16.1,17.8],
    'Costo Total':['$29000','$33000'],
    'Costo Total por Km':['$1801.25','$1853.94'],
    'Identificador':['Fucsia','Aqua']
}
df = pd.DataFrame(data
                  )
app.layout = dbc.Container([
    html.H1("TRABAJO FINAL TECNOTRANSPORTE", style={'text-align': 'center', 'color':'black'}),
    html.H2("Harold Stiven Rodriguez Rojas", style={'text-align': 'center', 'color':'black'}),
    html.H2("Código: 20221579052", style={'text-align': 'center', 'color':'black'}),
    html.H2("Universidad Distrital Francisco Jose de Caldas", style={'text-align': 'center', 'color':'black'}),
    html.H2("Facultad Tecnologica", style={'text-align': 'center', 'color':'black'}),
    html.H2("Teoria y Logica de programación", style={'text-align': 'center', 'color':'black'}),
    html.Hr(),
    html.H3("SELECCIÓN DE RUTAS ALTERNATIVAS", style={'text-align': 'center', 'color':'grey'}),
    html.Hr(),
    html.Iframe(id='map', srcDoc=map._repr_html_(), width='100%', height='500'),
    html.Div([
        dash_table.DataTable(
            columns=[
                {'id':'Ruta', 'name': 'Ruta'},
                {'id':'Identificador', 'name': 'Identificador'},
                {'id':'Longitud Sabana (Km)', 'name': 'Longitud Sabana (Km)'},
                {'id':'Longitud Vía (Km)','name':'Longitud Vía (Km)'},
                {'id':'PHD (UND)','name':'PHD (UND)'},
                {'id':'Longitud Total (Km)','name':'Longitud Total (Km)'},
                {'id':'Costo Total','name':'Costo Total'},
                {'id':'Costo Total por Km','name':'Costo Total por Km'},

            ],
            data=df.to_dict('records'),

        )
    ])
])

if __name__ == '__main__':
  app.run_server(mode='inline', height=600, width='80%')
```
